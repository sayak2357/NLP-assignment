```python
# -*- coding: utf-8 -*-
"""
Created on Fri Aug 10 21:37:46 2018

@author: sayak
"""

f = open("book.txt")
import operator
import re, string
def clearup(s, chars):
    return re.sub('[%s]' % chars, '', s).lower()

dict ={}
count = 0
#z= f.read()
for tem in f:
    
    word = tem.split()
    
    for words in word:
        #print words
        words = words.lower()
        words = words.lstrip()
        words = words.rstrip()
        words = clearup(words, string.punctuation+string.digits)
        if words in dict.keys():
            inc = dict[words]
            inc+=1
            dict[words] = inc
        else:
            dict[words]=1
sorted_d = sorted(dict.items(), key=operator.itemgetter(1))
des_sorted = sorted_d[::-1]   # sorted_d is a tuple 

types = 0
tokens = 0
prod=[]

for k,v in des_sorted:
    y=len(k)
   # print('product of freq & length = ',v*y)
    p=v*y
    prod.append(p)
xval=[]
for x in range(0,len(prod)):
    xval.append(x)
rank=[]
j=0
freq=[]
for k,v in des_sorted:
    j+=1
    rank.append(j)
    freq.append(v)
    
#print(des_sorted)    
#if words in dict.keys():
 #   types+=1
 #  tokens+= dict[words]
#print(xval)
import matplotlib.pyplot as plt
plt.scatter(xval,prod)
plt.title("Zip's Law for Length & Frequency")
plt.ylabel('Product of Length & Frequency')
plt.show()




```
