```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
f = open("alice.txt")
dict1={'test':1}
token_dict=0
```

```python
for sen in f:
    sentences = sent_tokenize(sen)
    for word in sentences:
        comp_words = word_tokenize(word)
        for cw in comp_words:
            cw=cw.lower()
            if cw.isalpha():
                if cw in dict1:
                    inc = dict1[cw]
                    inc+=1
                    dict1[cw] = inc
                    token_dict+=1
                else:
                    dict1[cw]=1
                    token_dict+=1
        
```

```python
#print(dict1)
f.close()
```

```python
f = open("alice.txt")
token_dict2=0
dict2={}
for sen in f:
    sentences = sent_tokenize(sen)
    for word in sentences:
        comp_words = word_tokenize(word)
        for i in range(len(comp_words)-1):
            if comp_words[i].isalpha():
                j=comp_words[i].lower()
                k=comp_words[i+1]
                l=j+' '+k
                if l in dict2:
                    inc = dict2[l]
                    inc+=1
                    dict2[l] = inc
                    token_dict2+=1
                else:
                    dict2[l]=1
                    token_dict2+=1
```

```python
#print(dict2['of anyone'])
f.close()
```
