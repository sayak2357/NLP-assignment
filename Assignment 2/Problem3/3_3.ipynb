```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
f = open("alice.txt")
dict1={'test':1}
token_dict=0
```

```python
for sen in f:
    sentences = sent_tokenize(sen)
    for word in sentences:
        comp_words = word_tokenize(word)
        for cw in comp_words:
            cw=cw.lower()
            if cw.isalpha():
                if cw in dict1:
                    inc = dict1[cw]
                    inc+=1
                    dict1[cw] = inc
                    token_dict+=1
                else:
                    dict1[cw]=1
                    token_dict+=1
        
```

```python
#print(dict1)
f.close()
```

```python
f = open("alice.txt")
token_dict2=0
dict2={}
for sen in f:
    sentences = sent_tokenize(sen)
    for word in sentences:
        comp_words = word_tokenize(word)
        for i in range(len(comp_words)-1):
            if comp_words[i].isalpha():
                j=comp_words[i].lower()
                k=comp_words[i+1]
                l=j+' '+k
                if l in dict2:
                    inc = dict2[l]
                    inc+=1
                    dict2[l] = inc
                    token_dict2+=1
                else:
                    dict2[l]=1
                    token_dict2+=1
```

```python
#print(dict2['of anyone'])
f.close()
```

```python
f = open("alice.txt")
token_dict3=0
dict3={}
for sen in f:
    sentences = sent_tokenize(sen)
    for word in sentences:
        comp_words = word_tokenize(word)
        for i in range(len(comp_words)-2):
            if comp_words[i].isalpha():
                j=comp_words[i].lower()
                k=comp_words[i+1]
                l=comp_words[i+2]
                m=j+' '+k+' '+l
                if m in dict3:
                    inc = dict3[m]
                    inc+=1
                    dict3[m] = inc
                    token_dict3+=1
                else:
                    dict3[m]=1
                    token_dict3+=1
```

```python
#print(dict3)
f.close()
```

```python
f = open("alice.txt")
token_dict4=0
dict4={}
for sen in f:
    sentences = sent_tokenize(sen)
    for word in sentences:
        comp_words = word_tokenize(word)
        for i in range(len(comp_words)-3):
            if comp_words[i].isalpha():
                j=comp_words[i].lower()
                k=comp_words[i+1]
                l=comp_words[i+2]
                m=comp_words[i+3]
                n=j+' '+k+' '+l + ' '+m
                if n in dict4:
                    inc = dict4[n]
                    inc+=1
                    dict4[n] = inc
                    token_dict4+=1
                else:
                    dict4[n]=1
                    token_dict4+=1
```

Print Bigram MLE

```python
for x in dict2:
    y=x.split()
    z=dict2[x]/dict1[y[0]]
    print(z)
```

MLE for Trigrams

```python
for x in dict3:
    y=x.split()
    z=dict3[x]/dict2[(y[0]+' '+y[1])]
    print(z)
    
```

MLE for Quadgrams

```python
for x in dict4:
    y=x.split()
    z=dict4[x]/dict3[(y[0]+' '+y[1]+' '+y[2])]
    print(z)
```
